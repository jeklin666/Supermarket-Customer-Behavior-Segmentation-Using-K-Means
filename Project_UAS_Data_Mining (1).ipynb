## Pra-pemrosesan dan Seleksi Fitur

### Subtask:
Clean and scale the numerical features for member customer segmentation.


```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# 1. Select relevant features
features = [
    'Unit price',
    'Quantity',
    'Sales',
    'Rating'
]
clustering_data = member_df[features].copy()

# 2. Ensure numeric types (handling potential string issues common in CSVs)
for col in features:
    if clustering_data[col].dtype == 'object':
        if col == 'Sales':
            # For 'Sales', remove thousands separators (dots) before converting to float
            clustering_data[col] = clustering_data[col].str.replace('.', '', regex=False).astype(float)
        else:
            # For other object columns, replace decimal comma with dot and convert to float
            clustering_data[col] = clustering_data[col].str.replace(',', '.', regex=False).astype(float)

# 3. Handle missing values if any
clustering_data = clustering_data.dropna()

# 4. Initialize and apply StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(clustering_data)

print("Shape of scaled features:", scaled_features.shape)
print("First 5 rows of scaled data:\n", scaled_features[:5])
```

## Penentuan Jumlah Cluster Optimal

### Subtask:
Determine the optimal number of clusters (k) for member segmentation using Elbow Method and Silhouette Analysis.


```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))

# Plotting results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Elbow Method
ax1.plot(K_range, inertia, 'bx-')
ax1.set_xlabel('Number of clusters (k)')
ax1.set_ylabel('Inertia')
ax1.set_title('Elbow Method for Optimal k')

# Silhouette Score
ax2.plot(K_range, silhouette_scores, 'ro-')
ax2.set_xlabel('Number of clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Analysis for Optimal k')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette_avg = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(refined_scaled_features)
    inertia.append(kmeans.inertia_)
    silhouette_avg.append(silhouette_score(refined_scaled_features, kmeans.labels_))

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Elbow Plot
ax1.plot(K_range, inertia, 'bo-', markersize=8)
ax1.set_xlabel('Number of Clusters (k)')
ax1.set_ylabel('Inertia (WCSS)')
ax1.set_title('Elbow Method for Refined Features')
ax1.grid(True)

# Silhouette Plot
ax2.plot(K_range, silhouette_avg, 'ro-', markersize=8)
ax2.set_xlabel('Number of Clusters (k)')
ax2.set_ylabel('Average Silhouette Score')
ax2.set_title('Silhouette Analysis for Refined Features')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

### Analisis Penentuan Klaster Optimal

Berdasarkan visualisasi di atas, berikut adalah interpretasi untuk menentukan jumlah klaster:

1.  **Metode Elbow**: Terlihat adanya 'siku' (elbow) atau penurunan inersia (WCSS) yang mulai melandai secara signifikan pada rentang k=4 hingga k=5. Ini menunjukkan bahwa penambahan klaster di atas 4 tidak lagi memberikan penurunan variansi internal yang drastis.
2.  **Analisis Silhouette**: Skor Silhouette rata-rata menunjukkan nilai yang cukup stabil dan kompetitif pada k=4 dibandingkan dengan nilai k lainnya (kecuali k=2 yang seringkali terlalu menyederhanakan data).

**Kesimpulan**: Mempertimbangkan keseimbangan antara kompleksitas model dan kemudahan interpretasi bisnis, kita menetapkan **k=4** sebagai jumlah klaster optimal untuk segmentasi member ini.

## PCA Implementation

### Subtask:
Apply Principal Component Analysis (PCA) to reduce the dimensionality of the refined scaled features to two components for 2D visualization.


**Reasoning**:
I will execute the PCA transformation and update the cleaned member dataframe with the resulting principal components.



```python
from sklearn.decomposition import PCA

# 1. Initialize and fit PCA
pca = PCA(n_components=2, random_state=42)
pca_result = pca.fit_transform(refined_scaled_features)

# 2. Add components to the clean dataframe
member_df_clean['PC1'] = pca_result[:, 0]
member_df_clean['PC2'] = pca_result[:, 1]

# 3. Print variance information
var_ratio = pca.explained_variance_ratio_
print(f"Explained Variance Ratio: PC1 = {var_ratio[0]*100:.2f}%, PC2 = {var_ratio[1]*100:.2f}%")
print(f"Total Variance Explained: {sum(var_ratio)*100:.2f}%")

display(member_df_clean[['Invoice ID', 'PC1', 'PC2']].head())
```
